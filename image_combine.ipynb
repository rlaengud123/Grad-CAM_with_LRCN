```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
from tqdm import tqdm, trange
import glob
import os
import cv2
```

```python
path = r"C:\Users\for_someday\Desktop\grad-cam-pytorch-master\results"
all_folder_list = os.listdir(r"C:\Users\for_someday\Desktop\grad-cam-pytorch-master\results")
save_path = r"C:\Users\for_someday\Desktop\grad-cam-pytorch-master\results2"

fig = plt.figure(figsize=[200,30]) # rows*cols 행렬의 i번째 subplot 생성
rows = 1
cols = 15

all_folder_list = ['lee5']

for k in tqdm(all_folder_list):
    folder_list = os.listdir(os.path.join(path, k))
    for j in trange(len(folder_list)):
        i = 1
        for filename in glob.glob(path + '/' + k + '/' + folder_list[j] + "/*.png"):
            img = cv2.imread(filename)
            ax = fig.add_subplot(rows, cols, i)
            ax.imshow(img)

            i += 1
        plt.savefig(save_path + '/' + k + '/' + folder_list[j] + '.png')
```

```python
path = r"F:\FDAI\성모병원\성모병원 코드\grad-cam-pytorch-master\results2"
all_folder_list = os.listdir(r"F:\FDAI\성모병원\성모병원 코드\grad-cam-pytorch-master\results2")
save_path = r"F:\FDAI\성모병원\성모병원 코드\grad-cam-pytorch-master\results3"


fig = plt.figure(figsize=[200,30*10]) # rows*cols 행렬의 i번째 subplot 생성
rows = 10
cols = 1
count = 0
all_folder_list = ['lee5']

start = 0
end = 10

for k in tqdm(all_folder_list):
    folder_list = os.listdir(os.path.join(path, k))
    file_list = list(glob.glob(path + '/' + k + "/*.png"))
    while(count<=56):
        i = 1
        for j in tqdm(file_list[start:end]):
            img = cv2.imread(j)
            ax = fig.add_subplot(rows, cols, i)
            ax.set_title(j.split('/')[1], fontsize=150)
            ax.imshow(img)
            i += 1
        plt.savefig(save_path + '/' + k + '/' + str(count) +'.png')
        count += 1
        start += 10
        end += 10
```
